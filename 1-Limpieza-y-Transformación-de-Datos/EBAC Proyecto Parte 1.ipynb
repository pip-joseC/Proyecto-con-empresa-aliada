# Unión de DataFrames

```python
df=sales.copy() #El dataframe unido será guardado en la variable df, en base a las ventas
df.head()
```

Podemos cambiar la columna de WEEK por una fecha con formato estandarizado, las cuales se encuentran en el DataFrame calendar

```python
dates=calendar[['WEEK','DATE']]
dates.head()
```

```python
def df_dict(dataframe,keys,values):
    dic={}
    for i in range(len(dataframe)):
        dic.update({dataframe.iloc[i][keys]: dataframe.iloc[i][values]}) 
    return dic

week_to_date=df_dict(dates,keys='WEEK',values='DATE') #Creating a dictionary where the keys are the weeks and the values are the dates
week_to_date
```

```python
week_sales=df['WEEK'].unique() #Estas son las únicas semanas que aparecen en las ventas
for week in week_sales:
    df.replace(week,week_to_date[week],inplace=True)
df.head()
```

El siguiente paso es juntar los dataframes de category y segment en el de product

Primero, anexaremos la columna de SEGMENT a product en base a lo indicado en el dataframe segment

```python
segment.head()
```

La columna SEGMENT depende de las columnas CATEGORY, ATTR1, ATTR2, ATTR3, FORMAT

```python
product=product.merge(segment, on=['CATEGORY','ATTR1','ATTR2','ATTR3','FORMAT'], validate='m:1') #Con la función merge, puedo juntar ambos dataframes en base a la columnas mencionadas de segment
product.head()
```

Ahora, juntamos product con category, reemplazando la columna de CATEGORY con el significado colocado en category

```python
category.head()
```

```python
id_to_category=df_dict(category,'ID_CATEGORY','CATEGORY') #Creating a dictionary where the keys are the ID and the values are the categories

for id in id_to_category.keys():
    product.replace(id,id_to_category[id],inplace=True)
product.head()
```
